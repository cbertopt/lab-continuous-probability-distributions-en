## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
# Define the range
lower_bound = 10  # Replace with your lower bound
upper_bound = 20  # Replace with your upper bound

# Generate uniform random numbers
num_samples = 5  # Specify the number of random numbers you want
random_numbers = uniform.rvs(loc=lower_bound, scale=upper_bound - lower_bound, size=num_samples)

print(random_numbers)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
def generate_uniform_random_numbers(bottom, ceiling, count):
    """
    Generate uniformly distributed random numbers.

    Parameters:
        bottom (float): The lower boundary of the generated numbers.
        ceiling (float): The upper boundary of the generated numbers.
        count (int): The number of random numbers to generate.

    Returns:
        numpy.ndarray: An array of uniformly distributed random numbers.
    """
    random_numbers = uniform.rvs(loc=bottom, scale=ceiling - bottom, size=count)
    return random_numbers
```

```python
# Call the function with the specified parameters
numbers1 = generate_uniform_random_numbers(bottom=10, ceiling=15, count=100)
numbers2 = generate_uniform_random_numbers(bottom=10, ceiling=60, count=1000)
```

```python
# Plotting the histograms
plt.figure(figsize=(12, 6))

# Histogram for the first set of parameters
plt.subplot(1, 2, 1)
sns.histplot(numbers1, bins=10, color='blue', kde=False)
plt.title('Uniform Distribution (10 to 15)')
plt.xlabel('Value')
plt.ylabel('Count')
plt.grid()

# Histogram for the second set of parameters
plt.subplot(1, 2, 2)
sns.histplot(numbers2, bins=10, color='green', kde=False)
plt.title('Uniform Distribution (10 to 60)')
plt.xlabel('Value')
plt.ylabel('Count')
plt.grid()

plt.tight_layout()
plt.show()
```

How are the two distributions different?

**Answer:** The key differences between the two distributions lie in their range, density, and overall appearance in the histogram. The first distribution is concentrated and narrow, while the second is broad and evenly distributed, highlighting the characteristics of uniform distributions based on their defined parameters.

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.

1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50

```python
from scipy.stats import norm 
```

```python
def generate_normal_random_numbers(mean, std_dev, count):
    """
    Generate normally distributed random numbers.

    Parameters:
        mean (float): The average of the generated numbers.
        std_dev (float): The standard deviation of the generated numbers.
        count (int): The number of random numbers to generate.

    Returns:
        numpy.ndarray: An array of normally distributed random numbers.
    """
    random_numbers = np.random.normal(loc=mean, scale=std_dev, size=count)
    return random_numbers
```

```python
# Generate 1,000 normally distributed numbers with mean=10 and std_dev=1
numbers1 = generate_normal_random_numbers(mean=10, std_dev=1, count=1000)
```

```python
# Generate 1,000 normally distributed numbers with mean=10 and std_dev=50
numbers2 = generate_normal_random_numbers(mean=10, std_dev=50, count=1000)
```

```python
# Plotting the distributions
plt.figure(figsize=(12, 6))

# Histogram for the first set of parameters
plt.subplot(1, 2, 1)
sns.histplot(numbers1, bins=30, color='blue', kde=True)
plt.title('Normal Distribution (mean=10, std_dev=1)')
plt.xlabel('Value')
plt.ylabel('Count')
plt.grid()

# Histogram for the second set of parameters
plt.subplot(1, 2, 2)
sns.histplot(numbers2, bins=30, color='green', kde=True)
plt.title('Normal Distribution (mean=10, std_dev=50)')
plt.xlabel('Value')
plt.ylabel('Count')
plt.grid()

plt.tight_layout()
plt.show()
```

How are the two distributions different?

**Answer:** The primary difference between the two distributions is the degree of variability represented by the standard deviation. A smaller standard deviation leads to a more concentrated distribution, while a larger standard deviation indicates a greater spread of values around the mean.

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
def generate_exponential_numbers(mean, size):
    """
    Generate exponentially distributed numbers.
    
    Parameters:
        mean (float): The mean of the distribution.
        size (int): The number of random numbers to generate.
    
    Returns:
        list: A list of exponentially distributed numbers.
    """
    # Calculate the rate parameter (lambda)
    rate = 1 / mean
    # Generate exponentially distributed numbers
    numbers = np.random.exponential(1 / rate, size)
    return numbers

# Generate two sequences
sequence_10 = generate_exponential_numbers(mean=10, size=10)
sequence_100 = generate_exponential_numbers(mean=10, size=100)

# Set the style for the plots
sns.set_style('whitegrid')

# Create histograms
plt.figure(figsize=(12, 6))

# Histogram for the sequence of 10
plt.subplot(1, 2, 1)
sns.histplot(sequence_10, bins=100, kde=True, color='blue')
plt.title('Exponential Distribution (Size 10)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Histogram for the sequence of 100
plt.subplot(1, 2, 2)
sns.histplot(sequence_100, bins=100, kde=True, color='orange')
plt.title('Exponential Distribution (Size 100)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.tight_layout()
plt.show()
```

How are the two distributions different?

**Answer:** The distribution generated from the sequence of size 100 will be smoother, more accurate, and better represent the characteristics of the exponential distribution compared to the sequence of size 10. This illustrates an important statistical principle: larger sample sizes generally provide better estimates of population parameters and more reliable insights into the underlying distribution characteristics.

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# Parameters
mean = 10  # mean time spent in minutes
lambda_param = 1 / mean  # rate parameter
x = 15  # time in minutes

# Calculate the probability P(X < x)
probability = 1 - np.exp(-lambda_param * x)

# Print the result
print(f"The probability that a customer will spend less than {x} minutes in the bank is: {probability:.4f}")
```
